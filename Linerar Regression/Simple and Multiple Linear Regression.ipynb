```python
### Creating our own class

class MeraLR1:
    def __init__(self):
        self.coef_= None
        self.intercept_ = None

    def fit(self,x_train,y_train):
        x_train = np.insert(x_train,0,1, axis=1) # Inserting 1 column at 0th index

        ## calculating the coeff
        betas = np.linalg.inv(np.dot(x_train.T,x_train)).dot(x_train.T).dot(y_train) # Beta = ((X^T.X)^-1).(X^T).(Y)
        self.coef_ = betas[1:]
        self.intercept_ = betas[0]

    def predict(self,x_test):
        y_pred = np.dot(x_test,self.coef_) + self.intercept_
        return y_pred
```

```python
lr1 = MeraLR1()
lr1.fit(x_train,y_train)
```

```python
lr1.intercept_ ,lr1.coef_
```

```python
y_pred = lr1.predict(x_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```
