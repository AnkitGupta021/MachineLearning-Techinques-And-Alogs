```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split


import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Encoding Target

lb = LabelEncoder()
df.species = lb.fit_transform(df.species)
df
```

```python
# Shuffling the data

df = df.sample(150)
```

```python
# Reseting the index
df.reset_index(drop=True, inplace=True)
df
```

```python
# visualiszation

sns.pairplot(df,hue='species')
```

```python
# Shorting the dataset by removing 0th class from target and considering only 2 predictor variables

new_df = df[df.species!=0][['sepal_length','sepal_width','species']] 
```

```python
new_df
```

```python
# Spliting the Data

x = new_df.iloc[:,:2]
y = new_df.iloc[:,-1]
```

```python
x.shape
```

```python
# Importing the models

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = KNeighborsClassifier()
clf3 = RandomForestClassifier()

esti = [('LR',clf1),('KNN',clf2),('RFC',clf3)]

# Using Cross Val Score to print the accuracy of each model before feeding them into voting classifier

for i in esti:
    x1= cross_val_score(i[1],x,y,cv=10,scoring='accuracy',n_jobs=-1)
    print(i[0], np.round(np.mean(x1),2))
```

```python
## Hard Voting

vc_hard = VotingClassifier(esti, voting='hard')
x2 = cross_val_score(vc_hard,x,y,cv=10, scoring='accuracy')
print(round(np.mean(x2),2))
                     
```

```python
## Soft Voting

vc_soft = VotingClassifier(esti, voting='soft')
x3 = cross_val_score(vc_soft,x,y,cv=10, scoring='accuracy')
print(round(np.mean(x3),2))
                     
```

```python
## Assigning different weights to the estimators with hard voting

for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vc_hard = VotingClassifier(esti, voting='hard',weights=[i,j,k])
            x2 = cross_val_score(vc_hard,x,y,cv=10, scoring='accuracy')
            print("For i={},j={},k={} :".format(i,j,k), round(np.mean(x2),2))
                     
```

### So max we are getting is 0.74 when weight of M1 i.e Logistic Regression is max and which makes send as individually it is giving us higher accuracy of 0.74

```python
## Assigning different weights to the estimators with soft voting

for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vc_soft = VotingClassifier(esti, voting='soft',weights=[i,j,k])
            x2 = cross_val_score(vc_soft,x,y,cv=10, scoring='accuracy')
            print("For i={},j={},k={} :".format(i,j,k), round(np.mean(x2),2))
                     
```

#### In soft voting too, so max we are getting is 0.69 when weight of M1 i.e Logistic Regression is max and which makes send as individually it is giving us higher accuracy of 0.74

```python
## Classifier with the same algo using SVM

from sklearn.svm import SVC

from sklearn.datasets import make_classification

# creating New Dataset

X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=2)

# Creating 5 model object with different parameters in terms of degree.

svm1 = SVC(probability=True, kernel='poly', degree=1)
svm2 = SVC(probability=True, kernel='poly', degree=2)
svm3 = SVC(probability=True, kernel='poly', degree=3)
svm4 = SVC(probability=True, kernel='poly', degree=4)
svm5 = SVC(probability=True, kernel='poly', degree=5)

estimators = [('svm1',svm1),('svm2',svm2),('svm3',svm3),('svm4',svm4),('svm5',svm5)]

# Calcuating the individual accuracy
for i in estimators:
    x1 = cross_val_score(i[1],X,y,cv=10,scoring='accuracy')
    print(i[0],np.round(np.mean(x1),2))
```

```python
# Feeding the same model in the voting classifier.

vc1 = VotingClassifier(estimators=estimators,voting='soft')
x = cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

### As we can see te he accuracy has been improved to 0.93, earlier max we were getting 0.89 so that can be used as an alternative to GridSearchCV() for hyperparameter tunnig.
