```python
import pandas as pd
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.DataFrame()
```

```python
df['X1'] = [1,2,3,4,5,6,6,7,9,9]
df['X2'] = [5,3,6,8,1,9,5,8,9,2]
df['label'] = [1,1,0,1,0,1,0,1,0,0]
```

```python
df
```

```python
import seaborn as sns
sns.scatterplot(x=df['X1'],y=df['X2'],hue=df['label'])
```

```python
df['weights'] = 1/df.shape[0]
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier(max_depth=1) # Decision Stump
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,2].values
```

```python
X,y
```

```python
# Step 2 - Train 1st model
dt1.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(dt1)
```

```python
# Plotting the decision boundaries

plot_decision_regions(X, y, clf=dt1, legend=2) 
```

```python
# Calculating y_pred for all the rows
df['y_pred'] = dt1.predict(X)
```

```python
df
```

```python
# Creating a function to calculate model weightage

def calculate_model_weight(error):

  return 0.5*np.log((1-error)/(error))
```

```python
# Step 3 - calculate model weight
# error =  total weight of misclassified rows, here we have 3 misclassified rows, so error = 0.1+0.1+0.1 = 0.3

alpha1 = calculate_model_weight(0.3)
alpha1
```

```python
# Step 4 - Creating function to Update weights 

def update_row_weights(row,alpha=0.423):
  if row['label'] == row['y_pred']:
    return row['weights'] * np.exp(-alpha)  # updated weight of correctly classified rows
  else:
    return row['weights'] * np.exp(alpha) # updated weight of misclassified rows.
```

```python
# Step5 = creating new column with updated weights

df['updated_weights'] = df.apply(update_row_weights,axis=1)
```

```python
df
```

```python
# checking sum of updated weight.

df['updated_weights'].sum()
```

```python
# Normalise -  Since sum of updated weight is not 1 we will normalise the weight to make it the sum=1 and create a new column

df['nomalized_weights'] = df['updated_weights']/df['updated_weights'].sum()
```

```python
df
```

```python
# Checkng sum of normalized weight, should be equal to 1

df['nomalized_weights'].sum()
```

```python
## Defining Range , we use np.cumsum() which return cumulative sum of the variable pass into it.
## Creating upper range

df['cumsum_upper'] = np.cumsum(df['nomalized_weights'])
```

```python
# Creating lower range 

df['cumsum_lower'] = df['cumsum_upper'] - df['nomalized_weights']
```

```python
df
```

```python
# Rearranging the columns 

df[['X1','X2','label','weights','y_pred','updated_weights','cumsum_lower','cumsum_upper']]
```

```python
df
```

```python
## Creating a functions for a new dataset that will return index of random values

def create_new_dataset(df):
    indices = []
    for i in range(df.shape[0]):
        a = np.random.random()
        for i in df.itertuples():
            if i.cumsum_upper > a and a > i.cumsum_lower:
                indices.append(i[0])
    return indices
```

```python
# Creating the random index (Upsampling)

index_values = create_new_dataset(df)

index_values
```

```python
# Creatig a new df

second_df = df.iloc[index_values,:3] # taking only first 4 columns and row as the index values
```

```python
second_df
```

```python
# Creating the m2 decision stump

dt2 = DecisionTreeClassifier(max_depth=1)
```

```python
# Splitting the data

X = second_df.iloc[:,0:2].values
y = second_df.iloc[:,2].values
```

```python
# Training the second DT

dt2.fit(X,y)
```

```python
plot_tree(dt2)
```

```python
# plotting the decision boundaries 

plot_decision_regions(X, y, clf=dt2, legend=2)
```

```python
# Calculating y_pred

second_df['y_pred'] = dt2.predict(X)
```

```python
second_df
```

```python
# Calculating alpha2, weightage to be given to m2, here error = 0.1 since only 1 class is misclassified

alpha2 = calculate_model_weight(0.1)
```

```python
alpha2
```

```python
# Step 4 - Update weights
def update_row_weights(row,alpha=1.09):
  if row['label'] == row['y_pred']:
    return row['weights'] * np.exp(-alpha)
  else:
    return row['weights'] * np.exp(alpha)
```

```python
# Updating the weights

second_df['updated_weights'] = second_df.apply(update_row_weights,axis=1)
```

```python
second_df
```

```python
# Checking the sum of updated weight should be 1
second_df.updated_weights.sum()
```

```python
# Normalising

second_df['nomalized_weights'] = second_df['updated_weights']/second_df['updated_weights'].sum()
```

```python
second_df
```

```python
# CHeckng sum of normalized weights

second_df['nomalized_weights'].sum()
```

```python
# Calculating range
# Upper limit

second_df['cumsum_upper'] = np.cumsum(second_df['nomalized_weights'])
```

```python
# Lower limit

second_df['cumsum_lower'] = second_df['cumsum_upper'] - second_df['nomalized_weights']
```

```python
# Rearranging

second_df[['X1','X2','label','weights','y_pred','nomalized_weights','cumsum_lower','cumsum_upper']]
```

```python
# Calculating index value for new dataframe

index_values = create_new_dataset(second_df)
```

```python
# New Df
third_df = second_df.iloc[index_values,[0,1,2,3]]
```

```python
third_df
```

```python
# Passing it to 3rd Decison Stamp 

dt3 = DecisionTreeClassifier(max_depth=1)

X = second_df.iloc[:,0:2].values
y = second_df.iloc[:,2].values

dt3.fit(X,y)
```

```python
plot_decision_regions(X, y, clf=dt3, legend=2)
```

```python
# Calculate y_pred

third_df['y_pred'] = dt3.predict(X)
```

```python
third_df
```

```python
# creating alpha 3, error = 0.7 as 7 are misclassified
alpha3 = calculate_model_weight(0.6)
alpha3
```
